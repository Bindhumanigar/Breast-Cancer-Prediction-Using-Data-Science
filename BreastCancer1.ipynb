# Data visualization

```python
#This graph represents the distribution of tumour stages among patients. Each slice of the pie chart corresponds to a
#tumour stage (Type 1, Type 2, Type 3, and Type 4), and the size of each slice represents the proportion of patients with 
#that tumour stage relative to the total number of patients.It's simply a visual style.

# Read the CSV file into a DataFrame
data = pd.read_csv('Data.csv')

# Create a pie chart using Matplotlib with a hole
plt.figure(figsize=(8, 8))
plt.pie(data['Tumour_Stage'].value_counts(),
        labels=['Type 1', 'Type 2', 'Type 3', 'Type 4'],
        autopct='%1.1f%%', startangle=90)
plt.title("Tumour Stages of Patients")
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle
plt.show()
```

```python
#This pie chart summarizes the distribution of surgery types among patients. 
#Each slice represents a different type of surgery, and the size of each slice corresponds to the proportion of patients
#who underwent that particular surgery type relative to the total number of patients. 
#The central hole is a design choice and doesn't convey additional information; it's simply a visual style.

# Get the value counts of the 'Surgery_type' column
surgery_counts = data['Surgery_type'].value_counts()

# Extract surgery types and their counts
transactions = surgery_counts.index
quantity = surgery_counts.values

# Create a pie chart using Matplotlib
plt.figure(figsize=(8, 8))
plt.pie(quantity, labels=transactions, autopct='%1.1f%%', startangle=90)
plt.title("Type of Surgery of Patients")
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle
plt.show()

```

```python
#This bar graph shows the distribution of histology types among patients. 
#Each bar represents a histology type, and its height indicates the count of patients with that histology type.

# Get the value counts of the 'Histology' column
histology_counts = data['Histology'].value_counts()

# Extract transaction names and quantities
transactions = histology_counts.index
quantity = histology_counts.values

# Create a bar graph using Matplotlib
plt.figure(figsize=(12, 6))
plt.bar(transactions, quantity, color='skyblue')
plt.xlabel('Histology')
plt.ylabel('Count')
plt.title('Histology of Patients')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.show()
```

```python
#This line graph illustrates the distribution of patient statuses across different age groups.
#The x-axis represents age groups, while the y-axis represents the count of patients. 
#Each line represents a different patient status (e.g., alive, deceased).

# Define the age bins and labels
age_bins = [0, 20, 40, 60, 80, 100]
age_labels = ['0-20', '20-40', '40-60', '60-80', '80-100']


# Bin the ages into groups
data['Age_Group'] = pd.cut(data['Age'], bins=age_bins, labels=age_labels, right=False)

# Create a line graph
plt.figure(figsize=(12, 6))
data.groupby(['Age_Group', 'Patient_Status']).size().unstack().plot(kind='line', marker='o')
plt.xlabel('Age Group')
plt.ylabel('Count')
plt.title('Patient Status by Age Group')
plt.grid(True)
plt.legend(title='Patient Status')
plt.show()
```

```python
#This histogram visualizes the distribution of ages among the patients.
#The x-axis represents age values, while the y-axis shows the frequency (count) of patients within each age group.

# Extract the 'Age' column from the DataFrame
sample = data["Age"]

# Create a histogram using Matplotlib
plt.hist(sample, rwidth=0.5,color="pink")

# Set the title and axis labels
plt.title("Distribution of Age")
plt.xlabel("Age")
plt.ylabel("Frequency")

# Display the histogram
plt.show()
```

```python
#This scatter plot compares the distribution of values for 'Tumour_Stage' and 'Histology' in the breast cancer dataset,
#showing the count of occurrences for each category of both variables. Each point represents a unique category, aiding in
#visualizing the relative frequencies of categories within each variable.

# Calculate value counts for 'Tumour_Stage'
tumour_stage_counts = data['Tumour_Stage'].value_counts()

# Calculate value counts for 'Histology'
histology_counts = data['Histology'].value_counts()

# Create scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(tumour_stage_counts.index, tumour_stage_counts.values, alpha=0.5, label='Tumour Stage')
plt.scatter(histology_counts.index, histology_counts.values, alpha=0.5, label='Histology',marker="*",color="red")
plt.title('Scatter Plot of Value Counts for Tumour Stage and Histology')
plt.xlabel('Value')
plt.ylabel('Count')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#This count plot displays the distribution of tumour stages among patients.
#Each bar represents a tumour stage, and the height of the bar indicates the count of patients with that particular tumour stage.

# Create a count plot using Seaborn to visualize the distribution of tumour stages
sns.countplot(x="Tumour_Stage", data=data)

# Set the title of the plot
plt.title('Tumour Stage of Patients')

# Display the plot
plt.show()
```

```python
#Box plot created,represents the distribution of values in the 'Protein1' column of the dataset.

# Choose a column for which you want to create a box plot, let's say 'Protein1'
sns.boxplot(x='Protein1', data=data)

# Set title
plt.title('Box Plot of Protein1')

# Show the plot
plt.show()
```

```python
# This visualization helps to understand the distribution of values within 'Protein1'.
#By this we can quickly understand the central tendency, spread, and skewness of the 'Protein1' data.

# Create a box plot with Seaborn
sns.boxplot(x=data["Tumour_Stage"], y=data["Protein1"], data=data)

# Set title
plt.title('Box Plot of Protein1 by Tumour Stage')

# Show the plot
plt.show()
```

```python
import numpy as np

# Assuming 'data' is a DataFrame containing the data
# Group the data by 'Tumour_Stage' and 'Histology' and calculate the count
grouped_df = data.groupby(['Tumour_Stage', 'Histology']).size().unstack(fill_value=0)

# Create a treemap-like visualization using imshow
plt.figure(figsize=(12, 6))

# Create a heatmap-like visualization
plt.imshow(grouped_df, cmap='Blues', aspect='auto', interpolation='nearest')

# Add text labels
for i in range(len(grouped_df.index)):
    for j in range(len(grouped_df.columns)):
        plt.text(j, i, grouped_df.iloc[i, j], ha='center', va='center', color='black')

# Customize ticks and labels
plt.xticks(np.arange(len(grouped_df.columns)), grouped_df.columns)
plt.yticks(np.arange(len(grouped_df.index)), grouped_df.index)
plt.xlabel('Histology')
plt.ylabel('Tumour Stage')
plt.title('Treemap-like Visualization of Tumour Stage and Histology')
plt.colorbar()
plt.show()
```

```python
#This heatmap illustrates the pairwise correlations between variables in the breast cancer dataset.
#Warm colors signify positive correlations, cool colors represent negative correlations, and darker shades indicate stronger 
#correlations. The annotated values in each cell display the correlation coefficient, ranging from -1 to 1.

# Drop 'Patient_ID' column as it's not relevant for correlation analysis
df_corr = data.drop(columns=['Patient_ID'])

# Compute the correlation matrix
corr_matrix = df_corr.corr()

# Create a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```
